# 1407. Top Travellers
[問題](https://leetcode.com/problems/top-travellers/description/?envType=study-plan-v2&envId=premium-sql-50)

## 解答の方針
ridesにusersをinner joinする。

```python
import pandas as pd
```

```python
data = [[1, 'Alice'], [2, 'Bob'], [3, 'Alex'], [4, 'Donald'], [7, 'Lee'], [13, 'Jonathan'], [19, 'Elvis']]
users = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 1, 120], [2, 2, 317], [3, 3, 222], [4, 7, 100], [5, 13, 312], [6, 19, 50], [7, 7, 120], [8, 19, 400], [9, 7, 230]]
rides = pd.DataFrame(data, columns=['id', 'user_id', 'distance']).astype({'id':'Int64', 'user_id':'Int64', 'distance':'Int64'})
```

```python
users
```

```python
rides
```

```python
def top_travellers(users: pd.DataFrame, rides: pd.DataFrame) -> pd.DataFrame:
    grouped = rides.groupby('user_id')['distance'].sum().reset_index()
    result = grouped.merge(users, left_on="user_id", right_on="id", how="outer")
    result["distance"] = result["distance"].fillna(0)
    result = result.dropna(subset="name")
    result = result.sort_values(by=["distance", "name"], ascending=[False, True])
    result = result.rename({"distance": "travelled_distance"}, axis=1)
    return result[["name", "travelled_distance"]]
```

```python
top_travellers(users, rides)
```
