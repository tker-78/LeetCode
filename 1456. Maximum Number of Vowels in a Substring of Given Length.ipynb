# 1456. Maximum Number of Vowels in a Substring of Given Length
[問題]()

Given a string s and an integer k, return the maximum number of vowel letters in any substring of s with length k.

Vowel letters in English are 'a', 'e', 'i', 'o', and 'u'.

## 解答の方針
sliding windowsで解答する。

## 実装

```python
from typing import List
class Solution(object):
    def maxVowels(self, s: str, k: int) -> int:
        ss = list(s)
        current_string = ss[:k]
        now = best = self.count_vowels(current_string)
        for i in range(1, len(ss)):
            current_string = ss[i:k+i] 
            now = self.count_vowels(current_string)
            if now > best:
                best = now
        return best


    
    def count_vowels(self, s: List[str]) -> int:
        vowels = ["a", "e", "i", "o", "u"]
        count = 0
        for i in s:
            if i in vowels:
                count += 1
        return count

```

```python
s = "abciiidef"
k = 3
# should be 3
s = "weallloveyou"
k = 7
```

```python
print(Solution().maxVowels(s, k))
```

## 改善
上記コードではTimeLimitExceededになる。
sliding windowで、文字列の割り当ては1回のみにして、
以降は、足し引きをするようにする

```python
class Solution2(object):
    def maxVowels(self, s: str, k: int) -> int:
        vowels = ["a", "e", "i", "o", "u"]

        # 最初のk文字の中にある母音をカウントする
        count = 0
        for i in range(k):
            if s[i] in vowels:
                count += 1
        best = count

        # 残りの部分の母音を反映する
        for i in range(k, len(s)):
            if s[i] in vowels:
                count += 1
            if s[i-k] in vowels:
                count -= 1
            if count > best:
                best = count 
        
        return best 


```

```python
s = ""
k = 3945 
# s = "abciiidef"
# k = 3
```

```python
print(Solution2().maxVowels(s, k))
```
