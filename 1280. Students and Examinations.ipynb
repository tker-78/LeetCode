## 解答の方針
examinationテーブルでグルーピングして、examの回数を算出する。  
studentsとsubjectsをcross joinする。  
最後に`fillna(0)`する。  


```python
import pandas as pd
```

```python
def students_and_examinations(students: pd.DataFrame, subjects: pd.DataFrame, examinations: pd.DataFrame) -> pd.DataFrame:
    grouped = examinations.groupby(["student_id", "subject_name"]).size().reset_index(name="attended_exams")
    student_subject = students.merge(subjects, how="cross")
    ans = student_subject.merge(grouped, on=["student_id", "subject_name"], how="left")
    ans.fillna(0, inplace=True)
    ans.sort_values(by=["student_id", "subject_name"], inplace=True)
    return ans
```

```python
students = """
| student_id | student_name |
| 1          | Alice        |
| 2          | Bob          |
| 13         | John         |
| 6          | Alex         |
"""
```

```python
subjects = """
| subject_name |
| Math         |
| Physics      |
| Programming  |
"""
```

```python
examinations = """
| student_id | subject_name |
| 1          | Math         |
| 1          | Physics      |
| 1          | Programming  |
| 2          | Programming  |
| 1          | Physics      |
| 1          | Math         |
| 13         | Math         |
| 13         | Programming  |
| 13         | Physics      |
| 2          | Math         |
| 1          | Math         |
"""
```

```python
students = students.replace(" ", "")
with open("data.txt", "w") as f:
    f.write(students)
students = pd.read_csv("data.txt", sep="|")
students.dropna(axis=1, inplace=True)
```

```python
subjects = subjects.replace(" ", "")
with open("data.txt", "w") as f:
    f.write(subjects)
subjects = pd.read_csv("data.txt", sep="|")
subjects.dropna(axis=1, inplace=True)
```

```python
examinations = examinations.replace(" ", "")
with open("data.txt", "w") as f:
    f.write(examinations)
examinations = pd.read_csv("data.txt", sep="|")
examinations.dropna(axis=1, inplace=True)
```
