# 1532. The Most Recent Three Orders

## 解答の方針
customer_idでグルーピングして、order_dateをdescending orderでソートする。  
`head(3)`で対象のレコードを取り出す。

```python
import pandas as pd
```

```python
data = [[1, 'Winston'], [2, 'Jonathan'], [3, 'Annabelle'], [4, 'Marwan'], [5, 'Khaled']]
customers = pd.DataFrame(data, columns=['customer_id', 'name']).astype({'customer_id':'Int64', 'name':'object'})
data = [[1, '2020-07-31', 1, 30], [2, '2020-7-30', 2, 40], [3, '2020-07-31', 3, 70], [4, '2020-07-29', 4, 100], [5, '2020-06-10', 1, 1010], [6, '2020-08-01', 2, 102], [7, '2020-08-01', 3, 111], [8, '2020-08-03', 1, 99], [9, '2020-08-07', 2, 32], [10, '2020-07-15', 1, 2]]
orders = pd.DataFrame(data, columns=['order_id', 'order_date', 'customer_id', 'cost']).astype({'order_id':'Int64', 'order_date':'datetime64[ns]', 'customer_id':'Int64', 'cost':'Int64'})
```

```python
customers
```

```python
orders
```

```python
def recent_three_orders(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    sorted_orders = orders.sort_values(by=["customer_id", "order_date"], ascending=[True, False])
    grouped = sorted_orders.groupby("customer_id")[["order_date", "order_id"]].apply(lambda x: x.head(3)).reset_index()
    merged = grouped.merge(customers, on="customer_id")
    result = merged[["name", "customer_id", "order_id", "order_date"]]
    result = result.rename({"name": "customer_name"}, axis=1)
    result = result.sort_values(by=["customer_name", "customer_id", "order_date"], ascending=[True, True, False])
    return result
```

```python
recent_three_orders(customers, orders)
```
