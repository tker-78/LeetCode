# 1571. Warehouse Manager

## 解答の方針
productsで掛け算をして、product_volumeを計算する。  
productsをwarehouseにouter joinする。  
warehouse内でvolumeを計算する。


```python
import pandas as pd
```

```python
data = [['LCHouse1', 1, 1], ['LCHouse1', 2, 10], ['LCHouse1', 3, 5], ['LCHouse2', 1, 2], ['LCHouse2', 2, 2], ['LCHouse3', 4, 1]]
warehouse = pd.DataFrame(data, columns=['name', 'product_id', 'units']).astype({'name':'object', 'product_id':'Int64', 'units':'Int64'})
data = [[1, 'LC-TV', 5, 50, 40], [2, 'LC-KeyChain', 5, 5, 5], [3, 'LC-Phone', 2, 10, 10], [4, 'LC-T-Shirt', 4, 10, 20]]
products = pd.DataFrame(data, columns=['product_id', 'product_name', 'Width', 'Length', 'Height']).astype({'product_id':'Int64', 'product_name':'object', 'Width':'Int64', 'Length':'Int64', 'Height':'Int64'})
```

```python
warehouse
```

```python
products
```

```python
def warehouse_manager(warehouse: pd.DataFrame, products: pd.DataFrame) -> pd.DataFrame:
    products["product_volume"] = products["Width"] * products["Length"] * products["Height"]
    merged = warehouse.merge(products, on="product_id", how="outer")
    merged["volume"] = merged["units"] * merged["product_volume"]
    grouped = merged.groupby("name")["volume"].sum().reset_index()
    grouped = grouped.rename({"name": "warehouse_name"}, axis=1)
    return grouped
```

```python
warehouse_manager(warehouse, products)
```
