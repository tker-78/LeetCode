# 1747. Leetflex Banned Accounts
[問題](https://leetcode.com/problems/leetflex-banned-accounts/description/?envType=study-plan-v2&envId=premium-sql-50)

## 解答の方針
account_idでグルーピングして、ログイン時間がクロスしているレコードがあるかチェックする。  
チェックする方法は、`df = df[(df['login_x'] <= df['logout_y']) & (df['login_y'] <= df['logout_x'])]`


```python
import pandas as pd
```

```python
data = [[1, 1, '2021-02-01 09:00:00', '2021-02-01 09:30:00'], [1, 2, '2021-02-01 08:00:00', '2021-02-01 11:30:00'], [2, 6, '2021-02-01 20:30:00', '2021-02-01 22:00:00'], [2, 7, '2021-02-02 20:30:00', '2021-02-02 22:00:00'], [3, 9, '2021-02-01 16:00:00', '2021-02-01 16:59:59'], [3, 13, '2021-02-01 17:00:00', '2021-02-01 17:59:59'], [4, 10, '2021-02-01 16:00:00', '2021-02-01 17:00:00'], [4, 11, '2021-02-01 17:00:00', '2021-02-01 17:59:59']]
log_info = pd.DataFrame(data, columns=['account_id', 'ip_address', 'login', 'logout']).astype({'account_id':'Int64', 'ip_address':'Int64', 'login':'datetime64[ns]', 'logout':'datetime64[ns]'})
```

```python
log_info
```

```python
def leetflex_banned_accounts(log_info: pd.DataFrame) -> pd.DataFrame:
    df = log_info.merge(log_info, how="cross")
    df = df[df["account_id_x"] == df["account_id_y"]]
    df = df[df["ip_address_x"] != df["ip_address_y"]]
    df = df[(df["login_x"] <= df["logout_y"]) & (df["login_y"] <= df["logout_x"])]
    df = df[["account_id_x"]]
    df = df.rename({"account_id_x": "account_id"}, axis=1)
    df = df.drop_duplicates()
    return df
```

```python
leetflex_banned_accounts(log_info)
```
