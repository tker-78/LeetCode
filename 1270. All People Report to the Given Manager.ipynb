# 1270. All People Report to the Given Manager
[問題](https://leetcode.com/problems/all-people-report-to-the-given-manager/description/?envType=study-plan-v2&envId=premium-sql-50)

## 解答の方針
manager_idにemployee_idを3回joinする。  
3回目のjoinのmanager_id == 1のレコードのみを返す。  
ただし、employe_id == 1は削除する。

```python
import pandas as pd
```

```python
data = [[1, 'Boss', 1], [3, 'Alice', 3], [2, 'Bob', 1], [4, 'Daniel', 2], [7, 'Luis', 4], [8, 'John', 3], [9, 'Angela', 8], [77, 'Robert', 1]]
employees = pd.DataFrame(data, columns=['employee_id', 'employee_name', 'manager_id']).astype({'employee_id':'Int64', 'employee_name':'object', 'manager_id':'Int64'})
```

```python
employees
```

```python
def find_reporting_people(employees: pd.DataFrame) -> pd.DataFrame:
    merged_1 = pd.merge(employees, employees, left_on="manager_id", right_on="employee_id")
    merged_2 = merged_1.merge(employees, left_on="manager_id_y", right_on="employee_id")
    merged_3 = merged_2.merge(employees, left_on="manager_id", right_on="employee_id", suffixes=["_z", "_r"])
    result = merged_3[["employee_id_x", "manager_id_x", "manager_id_y", "manager_id_z", "manager_id_r"]]
    result = result[result["manager_id_r"] == 1]
    result = result[~(result["employee_id_x"] == 1)]
    result = result.rename({"employee_id_x": "employee_id"}, axis=1)
    return result[["employee_id"]]
```

```python
find_reporting_people(employees)
```
